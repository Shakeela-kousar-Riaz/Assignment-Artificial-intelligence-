**7. Create a program that takes a student's numerical grade as input and prints a
corresponding letter grade. Use if-else statements to classify the grade into
categories such as "A," "B," "C," "D," or "F."**

o If the numerical grade is 90 or higher, the corresponding letter grade is "A."

o If the numerical grade is between 80 and 89 (inclusive), the corresponding
letter grade is "B."

o If the numerical grade is between 70 and 79 (inclusive), the corresponding
letter grade is "C."

o If the numerical grade is between 60 and 69 (inclusive), the corresponding
letter grade is "D."

o If the numerical grade is below 60, the corresponding letter grade is "F."

```python
#Get user input for numerical grade
numerical_grade =  float(input("Enter the student numerical grade:"))

#using if else statement determine the letter grade
if numerical_grade >= 90:
  letter_grade = "A"
elif 80 <= numerical_grade <= 89:
    letter_grade = "B"
elif 70 <= numerical_grade <= 79:
    letter_grade = "C"
elif 60 <= numerical_grade <= 69:
    letter_grade = "D"
else:
    letter_grade = "F"

#print the crossponding letter grade
print(f"The crosspnding letter grade is: {letter_grade}")
```
