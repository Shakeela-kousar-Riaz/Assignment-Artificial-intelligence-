Question 2: Perform the following task:

Explore the Netflix dataset to understand the distribution of movies vs. TV shows, identify the
countries producing the most content, and analyze the trend of releases over the years. What genres
are most common, and how does the rating distribution vary across genres?
Dataset: Netflix Shows

```python
full_data = pd.read_csv('/content/Netflix.zip')
```

```python
full_data.head(5)
```

```python

import pandas as pd

# Explore the distribution of movies vs. TV shows
content_distribution = full_data['type'].value_counts()

# Print the result
print("Distribution of movies vs. TV shows:")
print(content_distribution)
```

```python



# Identify the countries producing the most content
top_countries = full_data['country'].value_counts().head(5)

# Print the result
print("Top 5 countries producing the most content:")
print(top_countries)
```

```python


# Convert 'release_year' to datetime and extract the year
full_data['release_year'] = pd.to_datetime(full_data['release_year'], errors='coerce')
full_data['release_year'] = full_data['release_year'].dt.year

# Analyze the trend of releases over the years
release_trend = full_data['release_year'].value_counts().sort_index()

# Plot the trend
plt.figure(figsize=(12, 6))
release_trend.plot(kind='bar', color='skyblue')
plt.title('Netflix Content Releases Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python


# Split and stack the 'listed_in' column to extract genres
genres = full_data['listed_in'].str.split(', ', expand=True).stack()

# Get the count of each genre
common_genres = genres.value_counts().head(5)

# Print the result
print("Most common genres:")
print(common_genres)
```

```python
# Handle missing values
print("\nHandling missing values:")
print(fd.isnull().sum())

# Example: Impute missing ages with the median age
fd['Age'].fillna(fd['Age'].median(), inplace=True)
```

```python
# Create a new column indicating passenger age category
fd['Age_Category'] = pd.cut(fd['Age'], bins=[0, 18, 65, 100], labels=['Child', 'Adult', 'Elderly'])
```

```python
# Calculate the average fare by passenger class
average_fare_by_class = fd.groupby('Pclass')['Fare'].mean()

# Print the result
print("\nAverage Fare by Passenger Class:")
print(average_fare_by_class)
```
